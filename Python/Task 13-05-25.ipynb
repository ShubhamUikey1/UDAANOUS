```python
Data = pd.read_excel(r"D:\Intership\Python\12-05-25\heatmap_data.xlsx")
data1 = pd.DataFrame(Data)
data1
```

```python

numeric_columns = ['Days Present','Task Completion Rates','Interpersonal Skills Rating','Decision-Making Skills Rating']  # Example column names
num_row, num_col = 2, 2
fig, axes = plt.subplots(num_rows, num_cols, figsize=(10, 6))
axes = axes.flatten()  
for index, column in enumerate(numeric_columns):
    sns.kdeplot(data1[column], ax=axes[index],fill=True)
    axes[index].set_title(f'KDE Plot for {column}')

plt.tight_layout()
plt.show()
```

```python
from scipy.stats import shapiro
```

```python
numeric_columns = ['Days Present','Task Completion Rates','Interpersonal Skills Rating','Decision-Making Skills Rating']
shapiro_results = {}
for column in numeric_columns:
    stat, p_value = shapiro(data1[column])
    shapiro_results[column] = round(p_value, 3)
shapiro_results
```

```python
def sqrt_transformation(data1, column_name):
    data1[f'{column_name}_sqrt'] = np.sqrt(data1[column_name])
    stat, p_value = shapiro(data1[f'{column_name}_sqrt'])
    distribution = sns.kdeplot(data1[f'{column_name}_sqrt'], fill=True)
    plt.title('Distribution of √ Task Completion Rates')
    plt.show()
    return distribution, p_value

# Call the function
distribution, p_value = sqrt_transformation(data1, 'Task Completion Rates')
print(distribution)
print('p-value:', p_value)
```

```python
def log_transformation(data1, column_name):
    data1[f'{column_name}_log'] = np.log(data1[column_name])
    stat, p_value = shapiro(data1[f'{column_name}_log'])
    distribution = sns.kdeplot(data1[f'{column_name}_log'],fill= True)
    plt.title('Distribution of √ Task Completion Rates')
    plt.show()
    return distribution, p_value

# Call the function
distribution, p_value = log_transformation(data1, 'Task Completion Rates')
print(distribution)
print('p-value:', p_value)
```
