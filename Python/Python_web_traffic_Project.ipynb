# Step 1: Set Up Your Environment

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

#  Step 2: Import Libraries & Load Data

```python
Data = pd.read_csv(r"D:\Intership\Python\web_traffic_analysis_project\data\web_traffic.csv")
data1 = pd.DataFrame(Data)
data1
```

#  Step 3: Understand the Dataset

```python
data1.shape
```

# Step 4: Session Duration by Device

```python
sns.boxplot(x='Device', y='SessionDuration', data=data1)
plt.title("Session Duration by Device")
```

# Step 5: Pages Visited by User Type

```python
sns.barplot(x='UserType', y='PagesVisited', data=data1)
plt.title("Pages Visited by User Type")
```

# Step 6: Trend Over Time: Average Session Duration


```python
daily_trend = data1.groupby('Date')['SessionDuration'].mean()
daily_trend.plot(title='Average Session Duration Over Time')
```
