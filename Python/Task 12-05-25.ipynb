# Task 1 Using Heatmap Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
Data2= ['Task Completion Rates' , 'Sales Revenue Generated' , 'Decision-Making Skills Rating']
Heat_map=data1[Data2].corr().round(3)
plt.figure(figsize=(6, 6))
sns.heatmap(Heat_map, annot=True, cmap='Blues')  # you can change cmap if needed
plt.title("Task Completion Rates vs Sales Revenue Generated vs Decision-Making Skills Rating")
plt.show()
```

# Task 2 Using Boxplot Data

```python
Data = pd.read_excel(r"D:\Intership\Python\12-05-25\boxplot_data.xlsx")
Df = pd.DataFrame(Data)
Df
```

```python
Df.shape
```

```python
Df.columns
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x=Df["churned"],y=Df["Task Completion Rates"], color="skyblue")
plt.title("Task Completion Rates vs churned")
plt.show()
```

```python
churned_df = Df[Df['churned'] == 'Yes'] 
median_churned = churned_df['Task Completion Rates'].median()
print(f"Median Task Completion Rate for churned employees: {median_churned}")
```
