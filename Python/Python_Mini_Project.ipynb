# 	Load and clean sales data

```python
import pandas as pd

df = pd.read_csv("D:\Intership\Python\sales_analysis_project\data\sales.csv")

df.dropna(inplace=True)
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df
```

# 	Explore top products, monthly trends, regional sales.

```python
df['Month'] = df['OrderDate'].dt.month

# Top products
top_products = df.groupby('ProductName')['Sales'].sum().sort_values(ascending=False)

# Monthly sales
monthly_sales = df.groupby('Month')['Sales'].sum()

# Regional sales
regional_sales = df.groupby('Region')['Sales'].sum()

print("top_products",top_products)
print("monthly_sales",monthly_sales)
print("regional_sales",regional_sales)
```

# Create bar plots, line charts, and extract insights.

# Top 10 Products

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8,6))
top_products.head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Products by Sales')
plt.ylabel('Sales')
plt.show()
```

# Monthly Sales Trend

```python
plt.figure(figsize=(10,6))
monthly_sales.plot(kind='line', marker='o')
plt.title('Monthly Sales Trend')
plt.ylabel('Sales')
plt.xticks(range(1,13))
plt.grid(True)
plt.show()
```

# Regional Sales

```python
plt.figure(figsize=(8,5))
sns.barplot(x=regional_sales.index, y=regional_sales.values)
plt.title('Sales by Region')
plt.ylabel('Sales')
plt.xlabel('Region')
plt.xticks(rotation=45)
plt.show()
```
